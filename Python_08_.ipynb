#Ex 7

문제 4에서 만든 리스트 scores의 원소들의 분산을 계산하고 출력하세요.

```python
sub_variance=0
for x in range(length):
    sub_variance += ((scores[x]-mean)**2)

variance=    (1/((length)-1)) * (sub_variance)

print(variance)
```
